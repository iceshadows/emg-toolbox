```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
from emg_toolbox.features import extract_all_features
from emg_toolbox.filters import PlifEegSubtractFilter
```

```python
data = pd.read_csv("data/bb.csv")
xn = data["EMG1"].to_numpy()
baseline  = xn.mean(axis=0)
xn = xn - baseline
xn = signal.detrend(xn, axis=0, type='constant')
raw = xn
```

```python
fs = 2000 
lowcut = 10
highcut = 300
order = 4
nyquist = 0.5 * fs
low = lowcut / nyquist
high = highcut / nyquist
b, a = signal.butter(order, [low, high], btype='band')
filtered = signal.lfilter(b, a, xn)
xn = filtered
```

```python
filter = PlifEegSubtractFilter(sf=fs, pwlf=50)

# Apply the filter to your data
filtered_data = []
for data_point in xn:
    filtered_data.append(filter.run_subtract_filter(data_point))

# Reset the filter if needed
filter.reset_subtract_filter()
xn = filtered
```

```python
data = extract_all_features(data=xn,fs=fs,return_dataframe=True)
```
